```python
import matplotlib.pyplot as plt

# Define the points
points = [
    (8, 9), (4, 5),
    (8, 5), (4, 9), (10, 9),
    (9, 8), (9, 10),
    (8, 9), (0, 1),
    (0, 9), (8, 1)
]

# Split the points into x and y coordinates
x = [point[0] for point in points]
y = [point[1] for point in points]

# Plot the points
plt.scatter(x, y, marker='o', c='blue')
plt.title('Plot of Points')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.grid(True)
plt.show()
```
